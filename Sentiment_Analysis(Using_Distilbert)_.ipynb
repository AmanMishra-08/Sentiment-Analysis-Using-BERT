```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
# from google.colab import files
# uploaded = files.upload()
```

```python
# df = pd.read_csv("Tweets.csv")
reviews = [
    "I love this airline",
    "The service was terrible",
    "Flight was delayed",
    "Great job! The app crashed within five minutes",
    "The movie was not as bad as I expected",
     "I an so Upset With all the Good thing in life but not so Upset i should not be Disappointed about Good things "
]
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    "distilbert-base-uncased-finetuned-sst-2-english"
)

model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
#reviews = df["text"].tolist()[:10]
```

```python
for review in reviews:

    inputs = tokenizer(
        review,
        return_tensors="pt"
    )

    with torch.no_grad():
        outputs = model(**inputs)

    probs = torch.nn.functional.softmax(outputs.logits, dim=-1)

    prediction = torch.argmax(probs).item()

    if prediction == 1:
        label = "POSITIVE"
    else:
        label = "NEGATIVE"

    confidence = probs[0][prediction].item()

    print("\nReview:", review)
    print("Logits:", outputs.logits)
    print("Probabilities:", probs)
    print("Prediction Index:", prediction)
    print("Label:", label)
    print("Confidence:", confidence)
```
