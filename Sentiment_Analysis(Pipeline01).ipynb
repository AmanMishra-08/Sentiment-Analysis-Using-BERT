```python
import pandas as pd
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
reviews = [
    "I love this movie. It is amazing!",
    "This movie is terrible. I hate it.",
    "The acting was fantastic.",
    "Waste of time.",
    "The story was good but the ending was disappointing.",
    "I an so Upset With all the Good thing in life but not so Upset i should not be Disappointed about Good things ",
    "i am having fun in killing my cat"
]
```

```python
results = classifier(reviews)
```
